```python
# get gpe with modus
df = pd.read_csv(io.BytesIO(uploaded['uispraken.csv']))
count=0
locations = []
for index, row in df.iterrows():
  # pdb.set_trace()
  if pd.isna(row['Uitspraak']):
    continue
  try:
    doc = nlp_wk(row['Uitspraak'])
    # for each tenlastelegging, get the most occuring GPE
    local_list = []
    for ent in doc.ents:
      if ent.label_ == "GPE":
        local_list.append(ent.text)
    if local_list:
      # get the mode of the GPEs
      freqs = groupby(Counter(local_list).most_common(), lambda x:x[1])
      most_occurring_gpe = [val for val,count in next(freqs)[1]][0]
      if most_occurring_gpe == "Nederland":
        most_occurring_gpe = local_list[0]
      locations.append([row['ECLI'], row['Datum'], most_occurring_gpe])
    # if most_occurring_gpe == "Nederland":
    #   print(row['Uitspraak'])
  except ValueError:
    print(row['Uitspraak'])
    
```
